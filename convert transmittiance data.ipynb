# Preparing data in transmittance
This notebook follows similar methods of data preparation to the data organization normalization and resampling notebook. The major difference here is we started with files in transmittance and needed to convert them to Absorbance values. The relation ship between absorbance and transmittance is shown below.
$$
Absorbance = 2 - log(\%T)
$$



```python
import jcamp
import glob
import os
import numpy as np
import pandas as pd
import pubchempy as pcp
from getInchi import getInchi
import matplotlib.pyplot as plt
```

```python
#bath with the ir not in abs and wavenumbers
path = os.path.join(os.getcwd(), 'scrapedFTIR', 'ir','badUnits')
#jcamp extension  
extension = 'jdx'
all_files = glob.glob(os.path.join(path, "*.jdx")) #grab all jcamp files

parsedJcamps = [] #place to store the parsed jcamps
#use jcamp library to parse all the jcamps
for file in all_files:
    parsedJcamps.append(jcamp.jcamp_readfile(file))

tJcamps = [] #holds transmittance and wavenumber jcamps

#this looks through our parsed jcamps and only stores the ones not in micrometers and appends them to the tJcamps list
for jcamp in parsedJcamps:
    wavenumbers = jcamp.get('x_units', r'N\A').lower() != 'micrometers'
    if wavenumbers == True:
        tJcamps.append(jcamp)
```

```python
#create dataframe holding all of the parsed jcamps in transmittance and wavenumber
df = pd.DataFrame({'jcamp': tJcamps})
#add x arrays to dataframe
x_list = []
for i in df.jcamp:
    x_list.append(i['x'])
df['x'] = x_list
#add y values to dataframe
y_list = []
for i in df.jcamp:
    y_list.append(i['y'])
df['y'] = y_list

```

```python
df.head(7)
```

```python
#some of these are not in the proper wavenumbers scale, we will simply drop them if the x values are not what we are looking for
for i in range(0,len(df)):
    try:
        if df.x[i][0] < 200: #looking at the data the ones that start in the 2.0 range are not in the proper wavenumber scale
            df = df.drop(i)
    except:
        df = df.drop(i) #this is to drop empty sets of wavenumbers
```

```python
df = df.reset_index(drop=True) #reset the index with our remaining good values
df.head()
```

```python
# define the new x values you want to resample to
new_x = np.linspace(555, 3846, num=880)

#add new x values to dataframe
new_x_list = []
for i in range(0,len(df)):
    new_x_list.append(new_x)
df['x_'] = new_x_list
```

```python
#function to transform each element of y to whole number % not decimal
transform_func = lambda x: 100*x
#apply to y values
df['y'] = df['y'].apply(lambda x: [transform_func(e) for e in x])
#look at df
df.head()
```

```python
# define function to transform each element of the array
transform_func2 = lambda x: 2 - np.log10(x)

# apply function to each row of the "y" column and assign to new column "yA"
df['yA'] = df['y'].apply(lambda x: [transform_func2(e) for e in x])
```

```python
df.head()
```

```python
#use interpolation to calculate new y values for the resampled x values
y_ = []
for i in range(len(df)):
    try:
        y_.append(np.interp(new_x,df.x[i],df.yA[i],left=0))
    except:
        df=df.drop(i-1).reset_index(drop=True) #drop a row if there is an error to make it non interperable
df['y_']=y_ #add interpolated values to frame
    
```

```python
df.head()
```

```python
#normalize Y values
#hold normalized y values
y_n = []
#normalize the list of y values for each item
for i in df.y_:
    yn = [x/max(i) for x in i]
    y_n.append(yn)
df['y_n']=y_n
```

```python
plt.plot(df.x_[2],df.y_n[2]) #visualize a spectra after normalization
```

```python
#space to store inchi keys
inchi_keys = []
#get an inchi key for each molecule
for i in df.jcamp:
    try:
        inchi_keys.append(getInchi(i['cas registry no']))
    except:
        inchi_keys.append('')
#space to store smiles strings
smiles = []
#get a smile string for each molecule
for inchi in inchi_keys:
    try:
        result = pcp.get_compounds(inchi,'inchikey')
        smiles.append(result[0].canonical_smiles)
    except:
        smiles.append('')
df['smiles']=smiles
```

```python
mask = (df.smiles != '') #mask to select only items we were able to obtain smiles values for
```

```python
dfF = df[mask] #final dataframe where all expected values are accounted for
```
